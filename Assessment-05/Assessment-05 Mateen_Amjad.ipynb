```python
fruits_list = input('Enter three of your favorite fruits name: ')
fruits_list = fruits_list.split(',')
print(fruits_list)
```

```python
print('First fruit: ', fruits_list[0])
```

```python
print('Last fruit: ', fruits_list[-1])
```

```python
print('Second fruit: ', fruits_list[1])
```

```python
print('First two fruits, ', fruits_list[0:2])
```

```python
fruits_list.append('cherry')
print("After append:", fruits_list)
```

```python
fruits_list.pop(1)
print("After remove:", fruits_list)
```

```python
print("Length of fruits list:", len(fruits_list))
```

```python
numbers = input("Enter five numbers (spaces-separated): ")
num1, num2, num3, num4, num5 = numbers.split()
num1 = int(num1)
num2 = int(num2)
num3 = int(num3)
num4 = int(num4)
num5 = int(num5)
numbers_list = [num1, num2, num3, num4, num5]
print("Numbers lis: ", numbers_list)

sum = numbers_list[0] + numbers_list[1] + numbers_list[2] + numbers_list[3] + numbers_list[4]
print("Sum of your numbers", sum)
avg = sum/5
print("Average of your numbers", avg)
```
