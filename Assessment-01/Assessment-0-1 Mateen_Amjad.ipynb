## Question 1
Define Artificial Intelligence (AI). Explain how AI differs from Machine Learning and Deep Learning. Illustrate the relationship among these three concepts with examples.
## Answer:
Artificial Intelligence is a latest science technique that studies methods, research, and theories that extended human behaviour intelligence, and a machine can think like human brain and can take decisions after training without guiding. 

Artificial Intelligence is a broader concept. It is like to enable a machine to act and solve problem like human brain.
Example: Chatbots, youtube & social apps algorithms

Machine Learning is subset of AI, in which a model we trained using concepts like supervised and unsupervised learning. A model learn from training and can predict and take decision based on condition.
Example: Email filtering

Deep Learning is subset of ML and AI. It works on multi layered concept like NLP. It is for complex problems like self driving cars. It trains like human brain neurons. 
Example: Self driving cars

Weak AI is all which is developed yet. Like weak Ai includes Ai process which is make to solve a specific problems or tasks. It can work only on that problem or situation but not think beyond of that and other self decision which a human can take.
