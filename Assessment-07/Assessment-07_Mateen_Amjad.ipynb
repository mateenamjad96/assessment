### Python Programming Questions: for and while Loops

#### Python for Loop Questions

#### 1. Write a Python program to print numbers from 1 to 50 using a for loop.

```python
for i in range(1,51):
    print(i)
```

#### 2. Print all even numbers between 1 and 100 using a for loop.

```python
for i in range(1, 100):
    if i % 2 == 0:
        print(i)
```

#### 3. Write a program to calculate the sum of numbers from 1 to N entered by the user.

```python
N = int(input("Enter end number to print sum: "))
sum = 0
for i in range(1, N+1):
    sum += i
print(sum)
```

#### 4. Print the multiplication table of a number entered by the user (up to 10).

```python
table = int(input("Enter number for table: "))
for i in range(1, 11):
    print(f'{table} * {i} = {table*i}')
```

5. Write a program to reverse a string using a for loop.

```python
name = "Mateen"
reverse_string = ""
for i in name:
    reverse_string = i + reverse_string
print(reverse_string)
```

6. Count the number of vowels present in a string using a for loop.

```python
vowels = "aeiou"
count = 0
str_value = "orange"

for i in str_value:
    if i in vowels:
        count += 1
print(f'Number of vowels in {str_value} is: {count}')
```

7. Print the following pattern:
#### *
####  **
####  ***
####  ****
####  *****

```python
for i in range(1, 6):
    print("*" * i)
```

8. Write a program to check whether a number is prime using a for loop.

```python
num = int(input("Enter a number: "))
for i in range(2, num):
    if num % i == 0:
        print(num, "is not a prime number")
        break
else:
    print(num, "is prime number")
```

#### 9. Given a list:
numbers = [12, 5, 7, 18, 21, 30]

Print only numbers divisible by 3.

```python
numbers = [12, 5, 7, 18, 21, 30]
for i in numbers:
    if i % 3 == 0:
        print(i)
```

#### 10. Break and Continue
Print numbers from 1 to 20 but: - Skip multiples of 3. - Stop execution if number becomes greater than 15.

```python
for i in range(1, 21):
    if i % 3 == 0:
        continue
    if i > 15:
        break
    print(i)
```

#### 12. Calculate factorial of a number using a while loop.

```python
num = int(input("Enter a num: "))
factorial = 1
user_num = num

while num > 0:
    factorial *= num
    num -= 1
print(f'Factorial of {user_num} is {factorial}')
```
