### 1. Write a Python program that can do the following:
- You have 50 dollar
- You buy an item that is 15 dollar, that has a 3% tax
- Print how much money you have left, after purchasing the item.


```python
money = 50
item_price = 15
#3% tax 
tax = item_price * 0.03
total_cost = item_price + tax

money_left = money - total_cost

print("Total cost of item:", total_cost)
print("Money left after purchase:", money_left)
```

### 2. Ask the user how many days until their birthday
- Print an approx. number of weeks until their birthday
- 1 week is = to 7 days.


```python
days = int(input("How many days until your birthday? "))

weeks = days / 7

print(f"Approximate {weeks} weeks until your birthday.")
```

### 3. Create a variable â€œGradeâ€ holding an integer between 0 to100. Code to print the letter grade of the number grade variable.
Grades:
- A = 90 - 100
- B = 80 - 89
- C = 70-79
- D = 60 - 69
- F = 0 â€“ 59

```python
grade = int(input("Enter your grade (0-100): "))

if grade >= 90 and grade <= 100:
    print("Letter Grade: A")

elif grade >= 80 and grade <= 89:
    print("Letter Grade: B")

elif grade >= 70 and grade <= 79:
    print("Letter Grade: C")

elif grade >= 60 and grade <= 69:
    print("Letter Grade: D")

elif grade >= 0 and grade <= 59:
    print("Letter Grade: F")

else:
    print("Invalid grade entered")
```
